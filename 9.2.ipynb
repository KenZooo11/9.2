```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns


def count_seniority(row):
    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019, 9, 27)
    else:
        end_date = row['DateofTermination']
    seniority_days = (end_date - row['DateofHire']).days  
    seniority_years = seniority_days / 365.25  
    return seniority_years


df['DateofHire'] = pd.to_datetime(df['DateofHire'], format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], format='%m/%d/%y')
df['EmploymentDuration'] = (df['DateofTermination'] - df['DateofHire']).dt.days
df_duration = df[df['EmploymentDuration'].notna()]


df_duration = df_duration.copy()
df_duration['Seniority'] = df_duration.apply(count_seniority, axis=1)


median_duration = df_duration['Seniority'].median()


plt.figure(figsize=(12, 8))
sns.scatterplot(x='Seniority', y='RecruitmentSource', data=df_duration, alpha=0.6)
plt.title('StaÅ¼ pracy w zaleÅ¼noÅ›ci od ÅºrÃ³dÅ‚a rekrutacji')
plt.xlabel('StaÅ¼ pracy (lata)')
plt.ylabel('Å¹rÃ³dÅ‚o rekrutacji')
plt.grid(True)


plt.axvline(x=median_duration, color='red', linestyle='--')
plt.text(median_duration + 0.1, 0, f'Median duration: {median_duration:.2f} lata', color='red')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Seniority', y='RecruitmentSource', data=df_duration, hue='RecruitmentSource')
plt.title('StaÅ¼ pracy w zaleÅ¼noÅ›ci od ÅºrÃ³dÅ‚a rekrutacji')
plt.xlabel('StaÅ¼ pracy (lata)')
plt.ylabel('Å¹rÃ³dÅ‚o rekrutacji')


plt.axvline(x=median_duration, color='red', linestyle='--')
plt.text(median_duration + 0.1, 0, f'Median duration: {median_duration:.2f} lata', color='red')

plt.tight_layout()
plt.show()

```
